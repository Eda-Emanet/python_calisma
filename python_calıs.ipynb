# Aggregate ():
Bu fonk değişkenlere yapılacak istatistikleri bizim belirlemek istediğimiz durumlarda kullanılır. Aynı zmandanda farklı değişkenlere farklı istatistikleri yapmamızı sağlar.

```python
df = pd. DataFrame ({ "sınıf" : ["A","b","C"],
                        "degisken1" : [20,30,40],
                       "degisken2": [29,45,67]})
df
```

```python
df.groupby("sınıf")
```

```python
df.groupby("sınıf").aggregate([min,max, np.median])
```

Bu kısımda eğer istediğimiz istatistik ölçüsü pandasın içinde varsa tırnak işareti ile, ya da tırnak işaretsiz yoksa olduğu kütüphane ile birlikte yazılır np. median gibi

```python
df.groupby("sınıf").aggregate( { "degisken1": min,"degisken2" : max})
```

```python
df.groupby(["sex", "class"])[["survived"]].aggregate("mean").unstack()
```

# Bağımsız İki Örneklem T Testi (AB Testi):
İki grup ortalaması hakkında karşılaştırma yapmak için kullanılır.

iki varsayımı vardır normallik ve varyans homojenliği varsayımı

```python
import pandas as pd
a = pd.DataFrame([20,40,40,24])
b = pd.DataFrame([45,67,56,78])
a_b = pd.concat([a,b], axis =1)
a_b.columns = ["A","B"]
a_b.head()
```

Ama bu şekilde verilerin gözükmesimi istemeyiz yanlarda gruplanmış şekilde olmalı.

```python
import numpy as np
grup_a = np.arange(len(a))
grup_a = pd.DataFrame(grup_a)
grup_a[:] = "A"
a = pd.concat([a,grup_a], axis=1)
grup_b = np.arange(len(b))
grup_b = pd.DataFrame(grup_b)
grup_b[:]= "B"
b = pd.concat([b,grup_b], axis= 1)
ab= pd.concat([a,b])
ab.head()
```

 normallik varsayımı :

```python
from scipy.stats import shapiro
```

```python
shapiro(a_b.A)
```

burada h0 hipotezi reddedilemmeiş çünkü p-value 0.05 den ufak değil.


```python
shapiro(a_b.B)
```

Varyans homojenliği testi.
h0 = varyanslar homojendir
h1: varyanslar homojen değildir.

```python
import scipy.stats as stats
```

```python
stats.levene(a_b.A, a_b.B) # iki değişkende verilmelidir.
```

Burada p-value yine 0.05 den ufak olmadığı için h0 reddedilemez.

```python
stats.ttest_ind(a_b["A"],a_b["B"], equal_var= True)
```

burada değişkenleri sıralı şekilde vermeliyiz. equal_var ise varyansların homojenliğinin olup olmadığını belirtmek içindri. Üst kısımda test ettiğimiz için true yazarız. Aynı zamanda p-value 0.05 den ufak olduğu için h0 hipotezi reddedilir. Yani ana kitle ortalamları birbirine eşit değildir.

```python
test_istatistiği, pvalue = stats.ttest_ind(a_b["A"], a_b["B"], equal_var = True)
print("Test İstatistiği = %.4f, P-value = %.4f" % (test_istatistiği,pvalue))
```
